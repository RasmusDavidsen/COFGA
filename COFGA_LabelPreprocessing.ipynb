## Loading the data

```python
# loading the train.csv to find features for each training point
train = pd.read_csv('dataset/train.csv')
print("Training data shape:\t", train.shape)
```

### showing data

```python
# Restructuring data
train.replace(' ', '_', regex=False,inplace=True)  # changed from regex=True to regex=False
train.replace('/', '_', regex=False,inplace=True)   # changed from regex=True to regex=False
train[1:4].T
```

### Preparing data by applying one hot encoding

```python
# applying the one hot encoding for the nescessary classes using pandas built in function
train = pd.get_dummies(train, columns=["sub_class", "general_class", "color"])

# Defining target for train by dropping coordinates, imgage id and tag ID.
train_target = train.iloc[:,11:]

# For now replacing -1 with 0 
train_target = train_target.replace(-1, 0)

numberObservations = train_target.shape[0]

train_target[1:4].T

```
